```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

#create dataset for visualisations
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) # to create a figure and an axis
ax.plot(x,y1) #to create a first plot
ax.plot(x,y2) #to create a second plot 
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" ) #to create a label for a legend
ax.plot(x,y2, label = "Person 2" ) #to create a second label for a legend
ax.legend(loc='lower right') #to create a legend
fig.show()
```

```python
#to create a new dataset
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(['Luffy','Zoro','Nami', 'Usopp','Sanji'])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7,3)) #to create a vis with multiple axes in one figure

ax1.scatter(x1, y1)
ax2.bar(x2,y2)
```

```python
#to create a figure with multiple axes and place them in one layout. 
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], #to set the layout of the figure
                                    ['ax2','ax3']],
                                    figsize = (7,4),
                                    layout = 'constrained') #to set constrained layout
someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)
someaxes['ax1'].set_xlabel('A big label',fontsize=18)
someaxes['ax2'].set_xlabel('Another label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2: 2 fast 2 Furious', fontsize=18)
```

```python
#create a new dataset
x = np.array(['Luffy','Zoro','Nami', 'Usopp','Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

#to create a two different graphs on one axis
fig, ax = plt.subplots(figsize=(7, 3))

ax.bar(x, y1, color = 'indigo') 
ax.plot(x, y2, color = 'red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = 'red')
ax.errorbar(x, y2, #to create error bars on a visual
            yerr=y2_sd, 
            fmt = 'none',
            ecolor = 'green',
            elinewidth = 1, 
            capsize = 6, 
            capthick = 1,
            errorevery = 2)
```
