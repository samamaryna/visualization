```python
#to import libraries we will be working with: 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
import seaborn
```

```python
#to create data for our plots
np.random.seed(613) #to lock a random numbers
x = np.arange(50) #to create a range of 50 numbers
y = np.random.randint(0, 100, 50) #to create a random range of 50 numbers from 0 to 100

fir, ax = plt.subplots(figsize = (5,3)) #size in inches? 
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
#ax.bar(x,y) #for bar plot

ax.plot(x,y) #for line plot
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) #to change the graph to histogram
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time') #to set the title
ax.set_ylabel('Total growth')#to set the y label
ax.set_xlabel('Years since start')#to set the x label
```

```python
#to create dictionaries for fonts
font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

#to change the fonts and their location
ax.set_title('Total growth over time', fontdict=font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict=font2, loc = 'top')
ax.set_xlabel('Years since start', fontdict=font2, loc = 'right')
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,
           y,
           marker ='*', #to change the marker style and its color
           color = 'blue')

```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, 
        y, 
        marker = '*', 
        color = 'indigo',
        linestyle = '--', #to change the line style and width
        linewidth = '1')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, 
        y, 
        marker = '*',
        markersize = 12, 
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', #to use HEX codes for colors
        markerfacecolor = '#FFFFFF')

#to change the grid style and color
ax.grid(axis = 'y',
        color = '#E4E4E4',
        linestyle = 'dotted',
        linewidth = 1)
```

```python
#to try other visual code from the Python library

df = pd.DataFrame({'x': x, 'y': y})

seaborn.scatterplot(data=df, x='x', y='y', size='x', legend=False, sizes=(20, 2000))
# show the graph
plt.show()
```
